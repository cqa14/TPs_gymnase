```python
import numpy as np
```

```python
R = [0, 10, 20, 50]
U = [10, 0, 0, 0]
I = [0, 0, 0, 0]

R[0] = R[1] + R[2] + R[3]

I[0] = U[0]/R[0]
for i in range(1,4):
    I[i] = I[0]
    U[i] = R[i]*I[i]

print(R,"\n",U,"\n",I)

UT = [10, 1.259, 2.512, 6.23]
IT = [0.126, 0.126, 0.126, 0.126]

DU = [0, 0, 0, 0]
DI = [0, 0, 0, 0]

for i in range(0,4):
    DU[i] = np.abs(U[i]-UT[i])/U[i]*100
    DI[i] = np.abs(I[i]-IT[i])/I[i]*100

print("\n",DU,"\n",DI)
```
